# White Dwarfs and Binaries - Kevin Burdge

## Gravitational Waves

Gravitational waves are "basically" self propogating perturbations in the space-time metric, generated by the time evolution of the quadrupole moment of the mass distribution, in particular it's 3rd time derivative.

## White Dwarfs

White dwarfs are remanants of most main sequence stars. Anything under 8 solar masses, they are degenerate, their electrons occupy Fermi degenerate energy levels. Their composition tends to be either Helium core or core made of carbon and oxygen as a result of fusion.  

Something interesting about Helium core stars is they can be formed only through binary evolution (find out why).  

Useful relation (read more about this)

### $$P_{RLO} \approx \dfrac{0.43}{\sqrt{\rho}}$$

The period in which object overflows its Roche Lobe. Mass gets transfered through Lagrange point.

The above equation says the orbital period (in days) in which an object will overflow its Roche Lobe is approximately $\frac{0.43}{\text{sqare root of its density (given in}\frac{gm}{cm^3})}$

When two stars like Sun are born in a close orbit, they can leave behind compact binary white dwarfs. It is thought that when these systems merge, they can result in a Type 1a supernova, which are extremely important in measuring cosmological distances.

These close Binary systems containing white dwarfs will be the vast majority of what LISA detects.

# Gravitational Microlensing - Przemek Mroz

General relativity tells us massive objects like clusters of galaxy can warp spacetime. Images of object behind the clusters will be distorted.  
Duration of gravitational microlensing - 100 days (black holes), 20 days (stars), 0.1 days (Earth sized objects).


# Hunting for White Dwarf merger remnants - Illaria Caiazzo

## The typical white dwarf 

- Mass $\sim$ half the mass of Sun
- Radius $\sim$ that of the Earth
- Hydrogen dominated atmosphere 
- Carbon-oxygen core

Some are magnetic  

- $\sim20%$ of white dwarfs are observed to be magnetic
- Magnetic fields range from a few tens of $kG$ to almost $10^9\;G$

## Finding White Dwarfs

Before $GAIA$  

- White dwarfs identified through spectroscopy (SDSS)
- Or through multiple-filter photometry and reduced proper motions
- Biased and incomplete sample

After $GAIA$  

- Precise photometry and astrometry for billions of stars, including white dwarfs
- Now we can identify white dwarfs using color and magnitude

## How do you find a merger remnant?

- Massive
- Rapidly rotating
- Highly magnetized


# Intro to ML - Niharika Sravan

## Types of problems to solve in data science 

### Classification

- Star vs Galaxy catalog
    - Binary classification problem
- Object type from survey LCs
    - Multi-class classification

### Regression

- Photometric redshift
- Orbital period for CVs

### Clustering

- Mostly unsupervised
- Often used with dimensionality reduction methods
- Can be later extended to perform classification or outlier detection
- Examples
    - Galaxy groups
    - SN 1a spectral sub-classes

### Generative Modeling

- Given features $X$ can we generate $X'$
- e.g given example of faces can we generate new ones.
- Examples in Astro
    - Synthetic Galaxies
    - Physics informed synthetic RR Lyrae light curves

### Transfer Learning

- Given existing model for task $D$ can we solve related problem $D'$
- e.g use classifier trained to differentiate between cats and dogs to classify tigers and wolves
- Examples in astro
    - Forcasting Supernova light curve using similar Supernova without similarity labels
        - K-means on top of Xception penultimate layer encoding bypassing need for autoencoder
    - Important to consider pitfalls of domain adaption
        - Transferred weights from adversarial training to identify merging galaxies to force domain-invariant feature extraction

### Active/Reinforcement Learning

- Given limited dataset $D$ can we focus training where most constraining
- e.g. given set of proteins which ones do we select for drug development
- Classsically is Optimal Design of Experiments
- Useful when acquiring training data is expensive
- Examples in astro
    - Variable star classification to counter Malmquist bias
    - Choosing galaxies benefiting from citizen scientist input

### Novel applications: 

#### Out of distribution events

- In ZTF data
- In real-time for transients

#### Optimal resource allocation

- Transient spectra to improve classifiers
- Galaxy spectra for distances and cosmology

## Considerations

- Missing Data

- Outliers
